```python
import matplotlib.pyplot as plt
import numpy as np
import math
import scipy.interpolate as ip
from sklearn.linear_model import LinearRegression
```

```python
nextFit = np.genfromtxt("nextFitWaste.csv", delimiter=',')
firstFit = np.genfromtxt("firstFitWaste.csv",delimiter=',')
bestFit = np.genfromtxt("bestFitWaste.csv", delimiter=',')
firstFitDec = np.genfromtxt("firstFitDecWaste.csv",delimiter=',')
bestFitDec = np.genfromtxt("bestFitDecWaste.csv", delimiter=',')


def findMeanOfArrays(array):
    meanArray1 = []
    meanArray2 = []
    meanArray3 = []
    meanArray4 = []
    meanArray5 = []
    meanArray6 = []
    meanArray7 = []
    meanArray8 = []
    meanArray9 = []
    meanArray10 = []
    meanArray11 = []
    meanArray12 = []
    meanArray13 = []
    meanArray14 = []

    for i in range(array[:,0].size):
        if (array[i][0] == 10):
            meanArray1.append(array[i][1])
        if (array[i][0] == 20):
            meanArray2.append(array[i][1])
        if (array[i][0] == 50):
            meanArray3.append(array[i][1])
        if (array[i][0] == 100):
            meanArray4.append(array[i][1])
        if (array[i][0] == 200):
            meanArray5.append(array[i][1])
        if (array[i][0] == 500):
            meanArray6.append(array[i][1])
        if (array[i][0] == 1000):
            meanArray7.append(array[i][1])
        if (array[i][0] == 2000):
            meanArray8.append(array[i][1])
        if (array[i][0] == 5000):
            meanArray9.append(array[i][1])
        if (array[i][0] == 10000):
            meanArray10.append(array[i][1])
        if (array[i][0] == 20000):
            meanArray11.append(array[i][1])
        if (array[i][0] == 50000):
            meanArray12.append(array[i][1])
        if (array[i][0] == 100000):
            meanArray13.append(array[i][1])
        if (array[i][0] == 200000):
            meanArray14.append(array[i][1])

            
    mean1 = np.mean(meanArray1)
    mean2 = np.mean(meanArray2)
    mean3 = np.mean(meanArray3)
    mean4 = np.mean(meanArray4)
    mean5 = np.mean(meanArray5)
    mean6 = np.mean(meanArray6)
    mean7 = np.mean(meanArray7)
    mean8 = np.mean(meanArray8)
    mean9 = np.mean(meanArray9)
    mean10 = np.mean(meanArray10)
    mean11 = np.mean(meanArray11)
    mean12 = np.mean(meanArray12)
    mean13 = np.mean(meanArray13)
    mean14 = np.mean(meanArray14)
    
    
    return [mean1,mean2,mean3,mean4,mean5,mean6,mean7,mean8,mean9,mean10,mean11,mean12,mean13,mean14]

def findRegressionLine(xVal, yVal):
    reg = LinearRegression().fit(xVal.reshape(-1,1), yVal)
    return reg


xVal = [10,20,50,100,200,500,1000,2000,5000,10000,20000,50000,100000,200000]

yNextFit = findMeanOfArrays(nextFit)

logX = np.log2(xVal)
logNextFit = np.log2(yNextFit)
nextFitCoeffs = np.polyfit(logX, logNextFit, deg=1)
poly = np.poly1d(nextFitCoeffs)
yFit = lambda x: np.exp2(poly(np.log2(x)))


plt.scatter(xVal, yNextFit, c='b', label='nextFit')
plt.loglog(xVal, yNextFit, linestyle='', basex=2, basey=2)
plt.loglog(xVal, yFit(xVal), label='RegressionLine', basex=2, basey=2)
plt.xlabel('Array Size')
plt.ylabel('Waste (Excess Bins Created)')
plt.legend()
plt.show()
print("Regression Line Function: " + str(nextFitCoeffs[0]) + "x + " + str(nextFitCoeffs[1]) )

yFirstFit = findMeanOfArrays(firstFit)

logX = np.log2(xVal)
logFirstFit = np.log2(yFirstFit)
firstFitCoeffs = np.polyfit(logX, logFirstFit, deg=1)
poly = np.poly1d(firstFitCoeffs)
yFit = lambda x: np.exp2(poly(np.log2(x)))


plt.scatter(xVal, yFirstFit, c='b', label='firstFit')
plt.loglog(xVal, yFirstFit, linestyle='', basex=2, basey=2)
plt.loglog(xVal,yFit(xVal), label='RegressionLine', basex=2, basey=2)
plt.xlabel('Array Size')
plt.ylabel('Waste (Excess Bins Created)')
plt.legend()
plt.show()
print("Regression Line Function: " + str(firstFitCoeffs[0]) + "x + " + str(firstFitCoeffs[1]) )


yBestFit = findMeanOfArrays(bestFit)

logX = np.log2(xVal)
logBestFit = np.log2(yBestFit)
bestFitCoeffs = np.polyfit(logX, logBestFit, deg=1)
poly = np.poly1d(bestFitCoeffs)
yFit = lambda x: np.exp2(poly(np.log2(x)))

plt.scatter(xVal, yBestFit, c='b', label='bestFit')
plt.loglog(xVal, yBestFit, linestyle='', basex=2, basey=2)
plt.loglog(xVal, yFit(xVal), label='RegressionLine', basex=2, basey=2)
plt.xlabel('Array Size')
plt.ylabel('Waste (Excess Bins Created)')
plt.legend()
plt.show()
print("Regression Line Function: " + str(bestFitCoeffs[0]) + "x + " + str(bestFitCoeffs[1]) )


yFirstFitDec = findMeanOfArrays(firstFitDec)

logX = np.log2(xVal)
logFirstFitDec = np.log2(yFirstFitDec)
firstFitDecCoeffs = np.polyfit(logX, logFirstFitDec, deg=1)
poly = np.poly1d(firstFitDecCoeffs)
yFit = lambda x: np.exp2(poly(np.log2(x)))

plt.scatter(xVal, yFirstFitDec, c='b', label='firstFitDec')
plt.loglog(xVal, yFirstFitDec, linestyle='', basex=2, basey=2)
plt.loglog(xVal, yFit(xVal), label='RegressionLine', basex=2, basey=2)
plt.xlabel('Array Size')
plt.ylabel('Waste (Excess Bins Created)')
plt.legend()
plt.show()
print("Regression Line Function: " + str(firstFitDecCoeffs[0]) + "x + " + str(firstFitDecCoeffs[1]) )


yBestFitDec = findMeanOfArrays(bestFitDec)

logX = np.log2(xVal)
logBestFitDec = np.log2(yBestFitDec)
bestFitDecCoeffs = np.polyfit(logX, logBestFitDec, deg=1)
poly = np.poly1d(bestFitDecCoeffs)
yFit = lambda x: np.exp2(poly(np.log2(x)))

plt.scatter(xVal, yBestFitDec, c='b', label='bestFitDec')
plt.loglog(xVal, yBestFitDec, linestyle='', basex=2, basey=2)
plt.loglog(xVal, yFit(xVal), label='RegressionLine', basex=2, basey=2)
plt.xlabel('Array Size')
plt.ylabel('Waste (Excess Bins Created)')
plt.legend()
plt.show()
print("Regression Line Function: " + str(bestFitDecCoeffs[0]) + "x + " + str(bestFitDecCoeffs[1]) )
```
